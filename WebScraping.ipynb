```python
import requests
google_token = "AIzaSyBy6pJs4DApRua7iNF0T4DNSksHLoshyf8"
query = input()
url = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={query}&inputtype=textquery&fields=photos,formatted_address,name,geometry&key={google_token}"
page = requests.get(url)
if page.status_code == 200 : 
    data = page.json()
else : 
    print("Invalid Search or URL")

address = data["candidates"][0]["formatted_address"]
print("Your Search Result is Here ")
print("Addresss ---> ",address)
photo_ref = data["candidates"][0]["photos"][0]["photo_reference"]
img_url = f"https://maps.googleapis.com/maps/api/place/photo?maxwidth=400&photoreference={photo_ref}&key={google_token}"
image = requests.get(img_url).content
f = open('temp.jpeg','wb')
f.write(image)
f.close()
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(plt.imread('temp.jpeg'))
plt.show()
```
